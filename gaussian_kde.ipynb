```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/roldugina/data_analysis_projects/main/Homes%20for%20Sale%20and%20Real%20Estate.csv")
```

```python
def kde(x,x_space):
    kde = stats.gaussian_kde(x)
    evaluate=kde.evaluate(x_space)
    print (type(kde))
    return evaluate
```

```python
values=df['Price']
values_space=np.linspace(min(values),max(values),100)
values_pdf=kde(values,values_space)
```

```python
fig,ax=plt.subplots(figsize=(12, 6))
```

```python
ax.hist(values,alpha=0.5, color='#FF7F00',bins=10,rwidth=0.9,density=True)
ax.plot(values_space,values_pdf, color='steelblue')

ax.set(xlim=(min(values),max(values)),ylim=(0,max(values_pdf*1.1)),label='Price pdf')
ax.grid(color='#FF7F00', linestyle='--',alpha=0.3)
ax.set_xticks(np.linspace(min(values), max(values), 10))
ax.tick_params(axis='x',labelrotation=0)
ax.tick_params(axis='y',labelrotation=0)
ax.set_ylabel('price pdf')
ax.set_title("Price")
ax.legend()
ax.ticklabel_format(style='plain')
fig
```
