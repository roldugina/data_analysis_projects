```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/roldugina/freecodecamp_projects/main/adult.data.csv')
df
```

```python
lambda_01=lambda x: (x=='>50K').sum()
```

```python
def summary(x):
    result = {
        'total': x['salary'].count(),
        'high_salary': lambda_01(x['salary'])
    }
    return pd.Series(result)
```

```python
df_salary1=df.groupby('native-country').apply(summary)
```

```python
df_salary2=df.groupby('native-country').agg(total=('salary','count'),high_salary=('salary', lambda x: sum(x == '>50K')))
```

```python
df_salary1['pct']=df_salary1['high_salary']/df_salary1['total']*100
df_salary1['pct'].idxmax()
```
