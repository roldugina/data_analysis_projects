```python
import requests
req_url='https://en.wikipedia.org/wiki/The_Simpsons'
req=requests.get(req_url)
req
```

```python
df4=pd.read_html(req.text)
len(df4)
```

```python
df4[2]
```

```python
episodes_table=df4[2]
episodes_table.head(5)
```

```python
col=episodes_table.columns
```

```python
col_names=col.get_level_values(2)
col_names
```

```python
episodes_table.columns=col_names
episodes_table=episodes_table.set_index('Season')
episodes_table.head(5)


```
